**Iris Dataset Analysis:**

 Data Wrangling & Visualization with Seaborn

**Project Overview**

This project involves data wrangling, exploratory data analysis (EDA), and visualization of the famous Iris dataset using Python (Pandas, Seaborn, Matplotlib) in Jupyter Notebook. We'll explore the relationships between different flower species and their sepal/petal measurements.

**Step 1:** Setup & Data Loading

1**.1 Import Required Libraries**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

**1.2 Load the Dataset**

```python
# Load Iris dataset directly from Seaborn
iris = sns.load_dataset('iris')
iris.head()  # Display first 5 rows
```

**Step 2:** Data Wrangling (Cleaning & Preprocessing)

**2.1 Check Missing Values**

```python
iris.isnull().sum()
```

**2.2 Check Data Types & Statistics**

```python
iris.info()  # All columns are numerical except 'species' (categorical)
iris.describe()  # Statistical summary
```

**2.3 Convert 'species' to Numerical (Optional for ML)
python**


```python
iris['species'] = iris['species'].map({'setosa': 0, 'versicolor': 1, 'virginica': 2})
```

**Step 3:** Exploratory Data Analysis (EDA) with Seaborn

**3.1 Pairplot (Relationships Between Features)**

```python
sns.pairplot(iris, hue='species', palette='viridis')
plt.suptitle("Pairplot of Iris Features by Species", y=1.02)
plt.show()
```

**3.2 Boxplot (Distribution by Species)
python**

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=iris, x='species', y='petal_length', palette='viridis')
plt.title("Petal Length Distribution by Species")
plt.xlabel("Species (0=Setosa, 1=Versicolor, 2=Virginica)")
plt.ylabel("Petal Length (cm)")
plt.show()
```

**3.3 Violin Plot (Density & Distribution)**

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=iris, x='species', y='sepal_width', palette='viridis')
plt.title("Sepal Width Distribution by Species")
plt.xlabel("Species")
plt.ylabel("Sepal Width (cm)")
plt.show()
```

**3.4 Heatmap (Correlation Matrix)**

```python
plt.figure(figsize=(8, 6))
sns.heatmap(iris.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix of Iris Features")
plt.show()
```

**3.5 Scatter Plot (Petal vs. Sepal)**

```python
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species', palette='viridis')
plt.title("Petal Length vs. Petal Width by Species")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.legend(title='Species')
plt.show()
```
